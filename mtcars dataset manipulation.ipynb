```python
import pandas as pd
```

```python
df = pd.read_csv('mtcars.csv')
df
```

```python
best_mpg = df['mpg'].max()
df.loc[df['mpg'] == best_mpg]
```

```python
worst_mpg = df['mpg'].min()
df.loc[df['mpg'] == worst_mpg]
```

```python
best_hp = df['hp'].max()
df.loc[df['hp'] == best_hp]
```

```python
df['hp'].median()
```

```python
df.groupby(['am']).mean()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['mpg'])
plt.xlabel('mpg')
plt.ylabel('Number of Cars')
plt.title('Miles per Gallon Histogram')
```

```python
import seaborn as sns

# Create a boxplot
sns.boxplot(x=df['cyl'], y=df['mpg'])

# Add labels and title
plt.title('Boxplot of MPG for Each Cylinder Type')
plt.xlabel('Cylinders')
plt.ylabel('Miles Per Gallon (mpg)')

# Show the plot
plt.show()
```

```python
# Calculate the correlation using the corr() method
correlation = df['wt'].corr(df['mpg'])

print(f"Correlation between weight and mpg: {correlation}")
```
